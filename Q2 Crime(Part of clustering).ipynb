```python
#Q2
# Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
from sklearn.preprocessing import normalize
```

```python
a = pd.read_csv("crime_data.csv")
a 
```

```python
#Normalized data fuction
def norm_func(i):
    x=(i-i.min())/(i.max()-i.min())
    return(x)
```

```python
df_norm=norm_func(a.iloc[:,1:])
```

```python
df_norm
```

```python
# create clusters
hc=AgglomerativeClustering(n_clusters=3,affinity='euclidean',linkage='complete')
hc
```

```python
y_hc=hc.fit_predict(df_norm)
y_hc
```

```python
a['h_clusterid']=hc.labels_
a
```

# Kmeans

```python
from sklearn.cluster import KMeans
```

```python
# Elbow curv
wcss=[]
for i in range(1,11):
    kmeans=KMeans(n_clusters=i)
    kmeans.fit(df_norm)
    wcss.append(kmeans.inertia_)
plt.plot(range(1,11),wcss)
plt.title('Elbow curv')
plt.xlabel('number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
# selecting 4 clusters from above scree plot
model=KMeans(n_clusters=4)
model.fit(df_norm)
model.labels_
```

```python
x=pd.Series(model.labels_)
a['Clust']=x
a
```

```python
a.iloc[:,1:5].groupby(a.Clust).mean()
```

# DBSCAN

```python
#Import the libraries
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
```

```python
a.info()
```

```python
df=a.iloc[:,1:5]
```

```python
df
```

```python
stscaler=StandardScaler().fit(df.values)
x=stscaler.transform(df.values)
x
```

```python
from sklearn.neighbors import NearestNeighbors
```

```python
a.shape
```

```python
neigh = NearestNeighbors(n_neighbors=14)
nbrs = neigh.fit(x)
distances, indices = nbrs.kneighbors(x)
```

```python
distances = np.sort(distances, axis=0)
distances = distances[:,1]
plt.plot(distances)  
```

```python
dbscan=DBSCAN(eps=1.30,min_samples=5)
dbscan.fit(x)
```

```python

dbscan.labels_
```

```python
cl=pd.DataFrame(dbscan.labels_,columns=['cluster'])
```

```python
cl
```
